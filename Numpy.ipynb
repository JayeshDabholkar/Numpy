```python
#Build a 2D list with all random integers [0,255]
#10x10 greyscaler image

import random
a = [] #outer list

for x in range(10):
    b = [] #create empty list
    for y in range(10):
        i = random.randint(0,255)
        b.append(i)
    a.append(b)
for i in a:
  print(i)
```

```python
#Build a 2D list with all random integers [0,255]
#2160x3840 greyscaler image

import random
import time

st = time.time() 
#start
L = [] #outer list

for x in range(2160):
    b = [] #create empty list
    for y in range(3840):
        i = random.randint(0,255)
        b.append(i)
    L.append(b)
#end 

et = time.time()
print("The code took",et-st,"seconds to run")
```

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
birds = plt.imread("birds.jpeg")
```

```python
plt.imshow(birds, cmap = "gray")
```

```python
print(birds.ndim)
print(birds.shape)
print(birds.dtype)
```

```python
#cut bottom half of bird image

birds_copy = birds[:]
birds_copy.shape
start = 0
end = birds_copy.shape[0] // 2
step = 1
plt.imshow(birds_copy[start:end:step], cmap = 'gray')
```

```python
#cut top half of bird image

plt.imshow(birds_copy[end:], cmap = 'gray')
```

```python
#show left half of birds image
end = birds_copy.shape[1] // 2
plt.imshow(birds_copy[:, :end], cmap = 'gray')
```

```python
#show right half of birds image
end = birds_copy.shape[1] // 2
plt.imshow(birds_copy[:, end:], cmap = 'gray')
```

```python
#rotate the birds image to the right

#step 1 use transpose to convert rows into columns 

plt.imshow(birds_copy.T, cmap = 'gray')
```

```python
# step 2 reverse the rows to make it look rotated correctly

plt.imshow(birds_copy.T[::-1,], cmap = "gray")
```
